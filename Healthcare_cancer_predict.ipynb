```
import pandas as pd
import numpy as np
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.preprocessing import StandardScaler

```

```
train = pd.read_csv("healthcare_dataset.csv")
```

TRAIN TEST DATA


```
train_test_data = [train]

for dataset in train_test_data:
    dataset['Title'] = dataset['Medical Condition'].str.split().str[0]
```

COUNT THE DATA IN DATASET

```
train['Title'].value_counts()
```

UNWANTED DATA DROP


```
train_clean = train.drop(columns=["Name","Doctor","Hospital"],errors="ignore")
```

```
train_clean.T
```

TRAGET SET IN DATA FROM TH DATASET

```
train_clean["Target"] = train_clean['Medical Condition'].apply(lambda x: 1 if x.strip().lower() =="cancer" else 0)
```

```
train_clean["Target"]
```

AFTER TARGET SET AND MAPPING FROM THE VALUE TO DROP THE OLD COLUMN


```
  train_clean.drop("Medical Condition",axis=1 , inplace = True)
```

```
train_clean
```

```
train_clean.T
```

TO FIND THE SHAPE FROM DATASET


```

target = train['Medical Condition']
train_clean.shape , target.shape
```

FILL THE VALUE FROM STATISTIC FORMULA USE TO


```
train_clean["Age"] = train_clean["Age"].fillna(train_clean["Age"].median())
```

```
train_clean["Age"]
```

```
train_clean["Billing Amount"] = train_clean["Billing Amount"].fillna(train_clean["Billing Amount"].median())
```

```
train_clean["Billing Amount"]
```

```
for col in ["Gender","Blood Type","Admission Type","Test Result"]:
  if col in train_clean.columns:
    train_clean[col]=train_clean[col].fillna("unknow")
```

```
train_clean.loc[train_clean['Age'] <= 16, 'Age'] = 0
train_clean.loc[(train_clean['Age'] > 16) & (train_clean['Age'] <= 26), 'Age'] = 1
train_clean.loc[(train_clean['Age'] > 26) & (train_clean['Age'] <= 36), 'Age'] = 2
train_clean.loc[(train_clean['Age'] > 36) & (train_clean['Age'] <= 62), 'Age'] = 3
train_clean.loc[train_clean['Age'] > 62, 'Age'] = 4
```

```
train_clean.T
```

```
training_data = train_clean
x = training_data.drop("Target", axis=1)
y = training_data["Target"]

categorical_cols = x.select_dtypes(include=["object"]).columns
if len(categorical_cols)>0:
  x = pd.get_dummies(x , columns= categorical_cols,drop_first=True)
x_train ,x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled,y_train)
y_pred = knn.predict(x_test_scaled)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```
prediction = pd.DataFrame(data=x_test.values,columns=x_test.columns,index=x_test.index)
prediction["predicted_Target"] = y_pred
prediction["true_Target"] = y_test.values
```
